```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
master=pandas.read_csv("/Users/VyUng/mystuff/bm.csv")
data=pandas.read_csv("/Users/VyUng/mystuff/126.csv")
```

```python
date=data['date']
base=data['base']
#Take the log of base
log_base=[]
for i in base:
    log_base.append(math.log(i))
plt.scatter(date,log_base)
plt.show()
#Take 10% of the data to be test set and 90% to be training set
train_x = date[:-50]
test_x = date[-50:]
train_y = log_base[:-50]
test_y = log_base[-50:]
#Create linear regression object
regr=linear_model.LinearRegression()
#Train the model using the training sets
regr.fit(np.array(train_x).reshape(-1,1),train_y)
#Make predictions using the testing set
prediction=regr.predict(np.array(test_x).reshape(-1,1))
regr.score(np.array(test_x).reshape(-1,1),test_y)
```

```python
#Coefficients
print('Coefficients: \n', regr.coef_)
#Mean squared error
print("Mean squared error: %.2f" % mean_squared_error(test_y,prediction))
#Plot outputs: 
plt.scatter(test_x, test_y,  color='black')
plt.plot(test_x, prediction, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())
0
plt.show()
```
